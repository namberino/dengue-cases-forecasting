# Data splitting

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
patch_all()

connection_string = "mongodb_connection_string"
client = MongoClient(connection_string)
db = client["OpenDengue"]
collection = db["vietnam-national-level"]

dataset_vn = find_pandas_all(collection, {})

if '_id' in dataset_vn.columns:
    dataset_vn = dataset_vn.drop(columns=['_id'])

dataset_vn.head()
```

```python
dataset_vn['T_res'].unique()
```

```python
# load dataset
df = dataset_vn.copy()
df = df[df['T_res'] == 'Week']
df['calendar_start_date'] = pd.to_datetime(df['calendar_start_date'])
df = df.sort_values(by='calendar_start_date')
```

```python
# log transformation to stabilize variance
df['dengue_total_scaled'] = np.log1p(df['dengue_total'])

# normalize data
scaler = MinMaxScaler(feature_range=(0, 1))
df['dengue_total_scaled'] = scaler.fit_transform(df[['dengue_total_scaled']])
```

```python
df.head()
```

```python
df.shape
```

```python
# use past 12 time steps for forecasting
seq_len = 12
X, y = [], []
for i in range(len(df) - seq_len):
    X.append(df['dengue_total_scaled'].iloc[i:i+seq_len].values)
    y.append(df['dengue_total_scaled'].iloc[i+seq_len])

X, y = np.array(X), np.array(y)
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

# Random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
```

```python
plt.scatter(y_test, predictions, color='blue', alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('True values')
plt.ylabel('Predicted values')
plt.title('Confusion graph of Random Forest')
plt.show()
```

```python
# predict and inverse transform
y_pred = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred.reshape(-1, 1))
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# clip negative values
y_pred = np.maximum(y_pred, 0)
```

```python
plt.plot(df['calendar_start_date'].iloc[-len(y_test):], y_test, label="Actual cases")
plt.plot(df['calendar_start_date'].iloc[-len(y_pred):], y_pred, label="Predicted cases", linestyle='--')
plt.legend()
plt.title("Dengue feves cases predictions")
plt.xlabel("Date")
plt.ylabel("Number of cases")
plt.show()
```

```python
for i in range (-1, -50, -10):
    future_steps = 5 # forecast next 5 time steps
    last_sequence = X[i] # get last sequence from dataset
    future_predictions = []

    for _ in range(future_steps):
        next_pred = model.predict(last_sequence.reshape(1, seq_len))
        future_predictions.append(next_pred[0])

        # update sequence by removing first value and appending new prediction
        last_sequence = np.roll(last_sequence, -1)
        last_sequence[-1] = next_pred

    # convert predictions back to original scale
    future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1, 1))

    # generate future dates
    last_date = df['calendar_start_date'].iloc[i]
    future_dates = pd.date_range(start=last_date + pd.DateOffset(months=1), periods=future_steps, freq='M')

    plt.plot(df['calendar_start_date'].iloc[-len(y_test):], y_test, label="Actual cases")
    plt.plot(df['calendar_start_date'].iloc[-len(y_pred):], y_pred, label="Predicted cases", linestyle='--')
    plt.plot(future_dates, future_predictions, label="Forecasted cases", linestyle=':', color='red')
    plt.legend()
    plt.title("Cases forecasting")
    plt.xlabel("Date")
    plt.ylabel("Number of Cases")
    plt.show()
```

# Bidirectional LSTM model with Attention mechanism

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Bidirectional, Input, Attention
```

```python
seq_len = 24 # 24 time step sequence
X, y = [], []
for i in range(len(df) - seq_len):
    X.append(df['dengue_total_scaled'].iloc[i:i+seq_len].values)
    y.append(df['dengue_total_scaled'].iloc[i+seq_len])

X, y = np.array(X), np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))
```

```python
inputs = Input(shape=(seq_len, 1))
x = Bidirectional(LSTM(100, return_sequences=True, recurrent_dropout=0.2))(inputs)
x = Bidirectional(LSTM(100, return_sequences=True, recurrent_dropout=0.2))(x)

attention = Attention()([x, x])
x = tf.keras.layers.Concatenate()([x, attention])

x = LSTM(50, recurrent_dropout=0.2)(x)
x = Dropout(0.2)(x)
outputs = Dense(1)(x)

model = Model(inputs, outputs)
model.compile(optimizer='adam', loss='msle')
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), verbose=1)
```

```python
history.history.keys()
```

```python
plt.plot(history.history['loss'], label='Training loss', color='blue')
plt.plot(history.history['val_loss'], label='Validation loss', color='red')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and validation loss')
plt.legend()
plt.show()
```

```python
# predict & inverse transform
y_pred = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred.reshape(-1, 1))
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# clip negative predictions
y_pred = np.maximum(y_pred, 0)
```
